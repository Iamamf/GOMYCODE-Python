## Part 2: Data Exploration with ydata-Profiling

#### Use ydata-profiling to generate a report of the provided dataset.

##### The Tunisair Flight Dataset is too large with over 107 thousands rows so we have decided to work with a sample of 200 rows

```python
# Take a random sample of the determined size
sample_df = flight_df.sample(n=200, random_state=42)
```

```python
# Generate the profile report
# before generating a profile report, we are working with a sample to represent the whole. 

profile = ProfileReport(sample_df, title='Pandas Profiling Report')
```

```python
# Generate an HTML report
profile.to_file("sample_df.html")
```

```python
profile.to_notebook_iframe()
```

### 1. Missing Values
The report indicates:
- There are no missing cells in the dataset. The missing cells percentage is 0.0%.

### 2. Correlations
The correlation matrix shows the relationships between different columns:
- `Aircraft_code` has a significant positive correlation with `STATUS` (0.411).
- `Arrival delay` has some positive correlation with `Arrival point` (0.205) and `Departure point` (0.300).
- `STATUS` has a moderate positive correlation with `Arrival point` (0.428).

### 3. Outliers or Unusual Values
From the report:
- The `Arrival delay` column shows a maximum value of 1147, which is an outlier considering the 95th percentile is 239.05. The standard deviation is quite high at 115.21535, indicating significant variability.
- The column also has 74 zero values, representing 37.0% of the dataset, which might need further investigation to understand if these represent no delays or missing data.

### 4. Summary of Findings
#### Interesting Patterns and Correlations:
- **Correlation between `Aircraft_code` and `STATUS`**: This suggests that certain aircraft codes are associated with specific statuses, which could be due to operational patterns or aircraft conditions.
- **Correlation of `Arrival delay` with `Arrival point` and `Departure point`**: This could indicate that delays are more common for certain routes or airports, which might be due to traffic, weather conditions, or airport efficiency.
- **High variability in `Arrival delay`**: The dataset shows a significant range in arrival delays, with some extreme outliers. This high variability suggests that delays are not consistent and might be influenced by various unpredictable factors.

#### Issues and Challenges:
- **Outliers in `Arrival delay`**: The presence of extreme outliers and high standard deviation can affect analyses and might require further investigation to understand the reasons behind such delays.
- **Zero Values in `Arrival delay`**: A significant portion of the dataset (37.0%) has zero values for arrival delays. It is important to verify if these truly represent no delay or if there are recording issues.
